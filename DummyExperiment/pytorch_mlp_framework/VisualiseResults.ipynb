```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = "/home/s2677266/MLP/MLP_CW/DummyExperiment/pytorch_mlp_framework/PreTrainBaseModel001TH6HU24/result_outputs/summary.csv"
df = pd.read_csv(file_path)

# Create a figure
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot Train and Validation Loss on the left y-axis
ax1.plot(df['train_loss'], label='Train Loss', color='blue', linestyle=':')
ax1.plot(df['val_loss'], label='Validation Loss', color='cyan', linestyle=':')
ax1.set_xlabel("Epochs")
ax1.set_ylabel("Loss", color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Create a second y-axis for Accuracy and F1 Score
ax2 = ax1.twinx()
ax2.plot(df['train_acc'], label='Train Accuracy', color='red', linestyle='-')
ax2.plot(df['val_acc'], label='Validation Accuracy', color='orange', linestyle='-')
ax2.plot(df['train_f1'], label='Train F1 Score', color='green', linestyle='--')
ax2.plot(df['val_f1'], label='Validation F1 Score', color='purple', linestyle='--')
ax2.set_ylabel("Accuracy & F1 Score", color='black')
ax2.tick_params(axis='y', labelcolor='black')

# Legends
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Title and grid
plt.title("Plot of Loss, Accuracy, and F1 Score for Train and Validation Data")
ax1.grid()

# Show the plot
plt.show()
```
